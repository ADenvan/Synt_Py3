```python
# 4. __new__ / __cal__ / __del__ Sigleton
# __new__

class DataBase:

    __instance = None # Будет ссылкой на экземпляр.

    # cls -Сылается на текущий клас Point
    def __new__(cls, *args, **kwargs):
        if cls.__instance is None: # проверяем.
            cls.__instance =super().__new__(cls) # Создает новый экземпляр.

        return cls.__instance

    def __del__(self):
        # Если обект удален то Объект __instance сново принимать None
        DataBase.__instance = None

    # Каждей раз вызывая __init__ свойствы будут изменены.
    def __init__(self, user, psw, port):
        self.user = user
        self.psw =psw
        self.port =port

    def connect(self):
        print(f"сооединение с БД: {self.user}, {self.psw}, {self.port}")
    def close(self):
        print("закрыте соединения с БД")
    def read(self):
        return "данные из БД"
    def write(sefl, data):
        print(f"запись в БД {data}")

db_4 =DataBase('root', '1234', 80)
db_44 =DataBase('rbbot', '4321', 55)
```

```python
# 6. get_ set_
# _protected -
# __prvate -
from accessify import private, protected

class Point:
    def __init__(self, x=0, y=0):
        self.__x = self.__y = 0
        if self.__check_value(x) and self.__check_value(y):
            self.__x = x
            self.__y = y

    # Метод класса. + Приватный. Проверка на числа.
    @classmethod
    def __check_value(cls, x): # cls -Ссылка на класс Point.
        return type(x) in (int, float)

    def set_coord(self, x, y):
        # Проверка на числа с приватным методам класса.
        if self.__check_value(x) and self.__check_value(y):
            self.__x = x
            self.__y = y
        else:
            raise ValueError("Координаты должны быть числами")

    def get_coord(self):
        return self.__x, self.__y

pt = Point(1, 2)
pt.set_coord(10, 20)
print(pt.get_coord())
```
