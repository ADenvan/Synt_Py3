```python
# pd.read_csv() - Функция для считывание дфнныя.
# dataframe.to_csv() - Метод для записи дфннах в файл с раширением CSV.
```

```python
import pandas as pd
```

```python
# read_csv - четение фаила.
df = pd.read_csv("file/book_data.csv")
df
```

```python
# sep= - Аргумент. разделитель
# index_col=[user_typ] - Аргумент. Указывает, что столбец 'user_typ' должен быть использован в качестве индекса.
# usecols= - Аргумент. Указывает, какие столбцы должны быть считаны из файла.

def df_sep(file: str, idx_col=str):
    df_sep = pd.read_csv(file, 
                        sep=',',
                        index_col=['city'],
                        usecols=['user_id', 'city', 'category_name', 'price', 'activation_data', 'year','month'])
    return df_sep

if __name__ == '__main__':
    
    log = df_sep("file/book_data.csv")
    print(log.info())
```

```python
# nrows=2 - Аргумент. Указывает, что нужно считать только первые две строки из файла.
# usecols=['city'] - Аргумент. Указывает, что нужно считать только столбец 'city'.

file_csv = "file/book_data.csv"
df_nrows = pd.read_csv(file_csv, 
                        usecols=['city'],
                        nrows=2)
df_nrows
```

```python
# parse_dates=[['category_name', 'price']] - Аргумент. Указывает, что столбцы 'category_name' и 'price' должны быть объединены в один столбец с типом данных datetime.
# keep_date_col=True - Аргумент. Указывает, что нужно сохранить исходные столбцы 'category_name' и 'price' в DataFrame.
file_csv = "file/book_data.csv"
df_parse_datas  = pd.read_csv(file_csv, 
                        parse_dates=[['category_name', 'price']], 
                        keep_date_col=True)
df_parse_datas 
```
