```python
# ТЕГ---Рекурсия-факториал.

def matryshka(N):
    if N == 1:
        print("Матрёшка.")
    else:
        print("Верх Матрёшки N= ", N)
        matryshka(N-1)
        print("Низ матрешки N= ", N)
print(matryshka(5))


# -------------------------------------------------------------
# Факториал это число умноженое на самосебя.
def fact41(n):
    if n <= 0:
        return 1
    else:
        return n * fact41(n - 1)
print(f"fact > {fact41(5)}")
```

```python
# Перестановка чисел.
# Только для двоичной.
def gen_bin(M, prefix=""):
    if M == 0:
        print(prefix)
        return
    gen_bin(M-1, prefix + digit)
    gen_bin(M-1, prefix + "1")
    for digit in "0", "1":
        gen_bin(M-1, prefix+digit)
        gen_bin(M-1, prefix + "1")
gen_bin(3)


# Перестановка чисел.
# Для произвольной.
def generate_number(N:int, M:int, prefix=None):
    """Генериует все числа (С лидируещим незначищем нулем)
    в N-ричной системе счисления (N <= 10) длина M
    """
    prefix = prefix or []
    if M == 0:  # Мышка
        print(prefix)
        return
    for digit in range(N):
        prefix.append(digit)  # Удленяем префикс цифры.
        generate_number(N, M-1, prefix)  # Генерируем. N-Система исчесления.
        prefix.pop()  # Удаляем последнюю цифру что была для генерайии
generate_number(3, 3)

# -------------------------------------------------------------
```

```python
# ТЕГ---рекурсия-генерация
def find(number, A):
    """ Ишет number в А и возвращает True, есди такое есть False, если такое нет
    """
    flag = False
    for x in A:
        if number == x:
            flag = True
            break
        return False

def generate_permutations(N:int, M:int=-1, prefix=None):
    """Генерация всех перестоновак N чисел в M позициях с префиксам prefix """
    M = N if M == -1 else M  # По умолчанию N числа в N позициях.
    prefix = prefix or []
    if M == 0:
        print(prefix)
        # print(prefix[0], prefix[1], prefix[2], end="")
        return
    for number in range(1, N+1):
        if find(number, prefix):
            continue
        prefix.append(number)
        generate_permutations(N, M-1, prefix)
        prefix.pop()
generate_permutations(3, 3)
```

```python
# Сортировка.
arr = [2, 10, -1, 8, 5, 1, 6, 9, 0, 7, 3, 4]
def quick_sort(nums):
  less = []
  equal = []
  greater = []

  if len(nums) > 1: # array len greater than 1 the exe
    pivot = nums[len(nums) // 2] #pivot at mid point
    for i in nums:
      if i < pivot: less.append(i)
      elif i == pivot: equal.append(i)
      elif i > pivot: greater.append(i)
    return quick_sort(less) + equal + quick_sort(greater)
  else:
    return nums # else return array
quick_sort(arr)
```

```python
# Сортировка.
arr = [2, 10, -1, 8, 5, 1, 6, 9, 0, 7, 3, 4]
def quick_sort(array):
    if len(array) < 2:
        return array
    else:
        pivot = array[0]
        less = quick_sort([i for i in array if i < pivot])
        greater = quick_sort([i for i in array if i > pivot])
        return less + [pivot] + greater
quick_sort(arr)
```
