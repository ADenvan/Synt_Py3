```python
# –ó–∞–¥–∞—á–∞ 1.
# –ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –∫–æ—Ç–æ—Ä–∞—è –Ω–∞–π–¥—ë—Ç –≤—Å–µ —á–∏—Å–ª–∞, –∫—Ä–∞—Ç–Ω—ã–µ 7, –Ω–æ –Ω–µ –∫—Ä–∞—Ç–Ω—ã–µ 5, –≤ –¥–∏–∞–ø–∞–∑–æ–Ω–µ –æ—Ç 2000 –¥–æ 3200 (–≤–∫–ª—é—á–∏—Ç–µ–ª—å–Ω–æ).
# –†–µ–∑—É–ª—å—Ç–∞—Ç –¥–æ–ª–∂–µ–Ω –±—ã—Ç—å –≤—ã–≤–µ–¥–µ–Ω –≤ –æ–¥–Ω—É —Å—Ç—Ä–æ–∫—É, —á–∏—Å–ª–∞ —Ä–∞–∑–¥–µ–ª–µ–Ω—ã –∑–∞–ø—è—Ç—ã–º–∏.

# –ü–æ–¥—Å–∫–∞–∑–∫–∞:
# –ò—Å–ø–æ–ª—å–∑—É–π—Ç–µ —Ñ—É–Ω–∫—Ü–∏—é range(#start, #end) –∏ –æ–ø–µ—Ä–∞—Ç–æ—Ä % –¥–ª—è –¥–µ–ª–µ–Ω–∏—è –ø–æ –º–æ–¥—É–ª—é.
st = []
for i in range(2000, 3201):
  st.append(str(i))
log = ','.join(st)
print(log)
```

```python
# üß© –ó–∞–¥–∞—á–∞ 2
# –£—Ä–æ–≤–µ–Ω—å: 1 (–ù–∞—á–∞–ª—å–Ω—ã–π)

# –£—Å–ª–æ–≤–∏–µ:
# –ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É –¥–ª—è –≤—ã—á–∏—Å–ª–µ–Ω–∏—è —Ñ–∞–∫—Ç–æ—Ä–∏–∞–ª–∞ –∑–∞–¥–∞–Ω–Ω–æ–≥–æ —á–∏—Å–ª–∞.
# –î–æ–ø—É—Å—Ç–∏–º, –Ω–∞ –≤—Ö–æ–¥ –ø–æ–¥–∞—ë—Ç—Å—è —á–∏—Å–ª–æ: 8
# –†–µ–∑—É–ª—å—Ç–∞—Ç: 40320

# –ü–æ–¥—Å–∫–∞–∑–∫–∞:
# –í–≤–æ–¥ –ø—Ä–µ–¥–ø–æ–ª–∞–≥–∞–µ—Ç—Å—è —Å –∫–æ–Ω—Å–æ–ª–∏. –ò—Å–ø–æ–ª—å–∑—É–π—Ç–µ —Ä–µ–∫—É—Ä—Å–∏—é.

def fact(n):
  if n == 0:
    return  1
  return n *  fact(n-1)
log = fact(8)
print(log)
```

```python
# üß© –ó–∞–¥–∞—á–∞ 3
# –£—Ä–æ–≤–µ–Ω—å: 1 (–ù–∞—á–∞–ª—å–Ω—ã–π)

# –£—Å–ª–æ–≤–∏–µ:
# –ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –∫–æ—Ç–æ—Ä–∞—è —Å –∑–∞–¥–∞–Ω–Ω—ã–º —á–∏—Å–ª–æ–º n —Å–æ–∑–¥–∞—ë—Ç —Å–ª–æ–≤–∞—Ä—å –≤–∏–¥–∞ {i: i*i} –¥–ª—è –≤—Å–µ—Ö i –æ—Ç 1 –¥–æ n.
# –ü—Ä–∏–º–µ—Ä –≤–≤–æ–¥–∞: 8
# –†–µ–∑—É–ª—å—Ç–∞—Ç: {1: 1, 2: 4, ..., 8: 64}

# –ü–æ–¥—Å–∫–∞–∑–∫–∞:
# –ò—Å–ø–æ–ª—å–∑—É–π—Ç–µ dict() –∏ —Ü–∏–∫–ª for.


def multy(n):
  d = dict()
  for i in range(1, n + 1):
    d[i] = i * i
  return d

log = multy(8)
print(log)
```

```python
# üß© –ó–∞–¥–∞—á–∞ 4
# –£—Ä–æ–≤–µ–Ω—å: 1 (–ù–∞—á–∞–ª—å–Ω—ã–π)

# –£—Å–ª–æ–≤–∏–µ:
# –ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –∫–æ—Ç–æ—Ä–∞—è –ø—Ä–∏–Ω–∏–º–∞–µ—Ç –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç—å —á–∏—Å–µ–ª, —Ä–∞–∑–¥–µ–ª—ë–Ω–Ω—ã—Ö –∑–∞–ø—è—Ç—ã–º–∏, –∏ —Å–æ–∑–¥–∞—ë—Ç –∏–∑ –Ω–∏—Ö —Å–ø–∏—Å–æ–∫ –∏ –∫–æ—Ä—Ç–µ–∂.
# –ü—Ä–∏–º–µ—Ä –≤–≤–æ–¥–∞: 34,67,55,33,12,98
# –†–µ–∑—É–ª—å—Ç–∞—Ç:
# –°–ø–∏—Å–æ–∫: ['34', '67', '55', '33', '12', '98']
# –ö–æ—Ä—Ç–µ–∂: ('34', '67', '55', '33', '12', '98')

# –ü–æ–¥—Å–∫–∞–∑–∫–∞:
# –ò—Å–ø–æ–ª—å–∑—É–π—Ç–µ –º–µ—Ç–æ–¥—ã .split() –∏ tuple().

def split_tuple(st: str):
  l = st.split(",")
  t = tuple(l)
  
st = "34,67,55,33,12,98"
log = split_tuple(st)
print(log)
```

```python
# üß© –ó–∞–¥–∞—á–∞ 5
# –£—Ä–æ–≤–µ–Ω—å: 1 (–ù–∞—á–∞–ª—å–Ω—ã–π)

# –£—Å–ª–æ–≤–∏–µ:
# –°–æ–∑–¥–∞–π—Ç–µ –∫–ª–∞—Å—Å —Å –¥–≤—É–º—è –º–µ—Ç–æ–¥–∞–º–∏:

# getString() ‚Äî –ø—Ä–∏–Ω–∏–º–∞–µ—Ç —Å—Ç—Ä–æ–∫—É –∏–∑ –≤–≤–æ–¥–∞;

# printString() ‚Äî –≤—ã–≤–æ–¥–∏—Ç —Å—Ç—Ä–æ–∫—É –≤ –≤–µ—Ä—Ö–Ω–µ–º —Ä–µ–≥–∏—Å—Ç—Ä–µ.

# –ü—Ä–æ—Ç–µ—Å—Ç–∏—Ä—É–π—Ç–µ –º–µ—Ç–æ–¥—ã –≤ –∫–æ–¥–µ.

# –ü–æ–¥—Å–∫–∞–∑–∫–∞:
# –ò—Å–ø–æ–ª—å–∑—É–π—Ç–µ –º–µ—Ç–æ–¥ __init__ –¥–ª—è –∏–Ω–∏—Ü–∏–∞–ª–∏–∑–∞—Ü–∏–∏.

class InputOutString(object):
  def __init__(self):
    self.s = ""

  def getString(self, st):
    self.s = st

  def printString(self):
    print (self.s.upper())

strObj = InputOutString()
strObj.getString("console log")
strObj.printString()
```

```python
# –ó–ê–î–ê–ß–ê 6
# –£—Ä–æ–≤–µ–Ω—å: –°—Ä–µ–¥–Ω–∏–π

# –£—Å–ª–æ–≤–∏–µ:
# –ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –∫–æ—Ç–æ—Ä–∞—è –≤—ã—á–∏—Å–ª—è–µ—Ç –∑–Ω–∞—á–µ–Ω–∏—è –ø–æ —Ñ–æ—Ä–º—É–ª–µ:
# Q = ‚àö[(2 * C * D) / H]
# –ì–¥–µ:
# C = 50, H = 30 (–∑–∞–¥–∞–Ω—ã),
# D ‚Äî –≤–≤–æ–¥–∏–º—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è (—á–µ—Ä–µ–∑ –∑–∞–ø—è—Ç—É—é).

# –ü—Ä–∏–º–µ—Ä –≤–≤–æ–¥–∞: 100,150,180
# –†–µ–∑—É–ª—å—Ç–∞—Ç: 18,22,24

# –ü–æ–¥—Å–∫–∞–∑–∫–∞:
# –û–∫—Ä—É–≥–ª–∏—Ç–µ —Ä–µ–∑—É–ª—å—Ç–∞—Ç –¥–æ –±–ª–∏–∂–∞–π—à–µ–≥–æ —Ü–µ–ª–æ–≥–æ —á–∏—Å–ª–∞.

import math

def exemp(inp):
  c = 50
  h = 30
  value = []
  items = [x for x in inp.split(",")]
  for d in items:
    
    value.append(str(int(math.sqrt(2 * c * float(d) / h))))
  return value

log = exemp("100,150,180")
print(log)
```

```python
# –ó–ê–î–ê–ß–ê 7
# –£—Ä–æ–≤–µ–Ω—å: –°—Ä–µ–¥–Ω–∏–π

# –£—Å–ª–æ–≤–∏–µ:
# –ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –∫–æ—Ç–æ—Ä–∞—è –ø—Ä–∏–Ω–∏–º–∞–µ—Ç 2 —á–∏—Å–ª–∞ X, Y –∏ –≥–µ–Ω–µ—Ä–∏—Ä—É–µ—Ç –¥–≤—É–º–µ—Ä–Ω—ã–π –º–∞—Å—Å–∏–≤, –≤ –∫–æ—Ç–æ—Ä–æ–º –∑–Ω–∞—á–µ–Ω–∏–µ –ø–æ –∏–Ω–¥–µ–∫—Å—É i,j —Ä–∞–≤–Ω–æ i*j.

# –ü—Ä–∏–º–µ—Ä –≤–≤–æ–¥–∞: "3,5"
# –†–µ–∑—É–ª—å—Ç–∞—Ç: [[0, 0, 0, 0, 0], [0, 1, 2, 3, 4], [0, 2, 4, 6, 8]]

# –ü–æ–¥—Å–∫–∞–∑–∫–∞:
# –ò—Å–ø–æ–ª—å–∑—É–π—Ç–µ –≤–ª–æ–∂–µ–Ω–Ω—ã–µ —Ü–∏–∫–ª—ã –∏ –≥–µ–Ω–µ—Ä–∞—Ç–æ—Ä—ã —Å–ø–∏—Å–∫–æ–≤.

input_str = "3,5"
dimensions = [int(x) for x in input_str.split(',')]
rowNum = dimensions[0]
colNum = dimensions[1]
multilist = [[0 for col in range(colNum)] for row in range(rowNum)]
for row in range(rowNum):
    for col in range(colNum):
        multilist[row][col] = row * col
print (multilist)
```

```python
# –ó–ê–î–ê–ß–ê 10
# –£—Ä–æ–≤–µ–Ω—å: –°—Ä–µ–¥–Ω–∏–π

# –£—Å–ª–æ–≤–∏–µ:
# –ü—Ä–æ–≥—Ä–∞–º–º–∞ –ø—Ä–∏–Ω–∏–º–∞–µ—Ç —Å—Ç—Ä–æ–∫—É —Å–ª–æ–≤, —Ä–∞–∑–¥–µ–ª—ë–Ω–Ω—ã—Ö –ø—Ä–æ–±–µ–ª–∞–º–∏. –ù—É–∂–Ω–æ —É–¥–∞–ª–∏—Ç—å –¥—É–±–ª–∏–∫–∞—Ç—ã –∏ –æ—Ç—Å–æ—Ä—Ç–∏—Ä–æ–≤–∞—Ç—å —Ä–µ–∑—É–ª—å—Ç–∞—Ç –ø–æ –∞–ª—Ñ–∞–≤–∏—Ç—É.

# –ü—Ä–∏–º–µ—Ä –≤–≤–æ–¥–∞: hello world and practice makes perfect and hello world again
# –†–µ–∑—É–ª—å—Ç–∞—Ç: again and hello makes perfect practice world

# –ü–æ–¥—Å–∫–∞–∑–∫–∞:
# –ò—Å–ø–æ–ª—å–∑—É–π—Ç–µ set() –∏ sorted().

w = "hello world and practice makes perfect and hello world again"
words = [word for word in w.split(" ")]
print (" ".join(sorted(list(set(words)))))
```

```python
# –ó–ê–î–ê–ß–ê 11
# –£—Ä–æ–≤–µ–Ω—å: –°—Ä–µ–¥–Ω–∏–π

# –£—Å–ª–æ–≤–∏–µ:
# –ü—Ä–æ–≥—Ä–∞–º–º–∞ –ø—Ä–∏–Ω–∏–º–∞–µ—Ç 4-–±–∏—Ç–Ω—ã–µ –±–∏–Ω–∞—Ä–Ω—ã–µ —á–∏—Å–ª–∞, —Ä–∞–∑–¥–µ–ª—ë–Ω–Ω—ã–µ –∑–∞–ø—è—Ç–æ–π, –∏ –≤—ã–≤–æ–¥–∏—Ç —Ç–æ–ª—å–∫–æ —Ç–µ, –∫–æ—Ç–æ—Ä—ã–µ –¥–µ–ª—è—Ç—Å—è –Ω–∞ 5.

# –ü—Ä–∏–º–µ—Ä –≤–≤–æ–¥–∞: 0100,0011,1010,1001
# –†–µ–∑—É–ª—å—Ç–∞—Ç: 1010

st_bin = "0100,0011,1010,1001"

value = []
items = [x for x in st_bin.split(',')]
for p in items:
    intp = int(p, 2)
    if not intp % 5:
        value.append(p)
print (','.join(value))
```

```python
# –ó–ê–î–ê–ß–ê 12
# –£—Ä–æ–≤–µ–Ω—å: –°—Ä–µ–¥–Ω–∏–π

# –£—Å–ª–æ–≤–∏–µ:
# –ù–∞–π—Ç–∏ –≤—Å–µ —á–∏—Å–ª–∞ –æ—Ç 1000 –¥–æ 3000 (–≤–∫–ª—é—á–∏—Ç–µ–ª—å–Ω–æ), —É –∫–æ—Ç–æ—Ä—ã—Ö –≤—Å–µ —Ü–∏—Ñ—Ä—ã ‚Äî —á—ë—Ç–Ω—ã–µ.

values = []
for i in range(1000, 3001):
    s = str(i)
    if (int(s[0]) % 2 == 0) and (int(s[1]) % 2 == 0) and (int(s[2]) % 2 == 0) and (int(s[3]) % 2 == 0):
        values.append(s)
print (",".join(values))
```

```python
# –ó–ê–î–ê–ß–ê 13
# –£—Ä–æ–≤–µ–Ω—å: –°—Ä–µ–¥–Ω–∏–π

# –£—Å–ª–æ–≤–∏–µ:
# –ü—Ä–æ–≥—Ä–∞–º–º–∞ –ø—Ä–∏–Ω–∏–º–∞–µ—Ç —Å—Ç—Ä–æ–∫—É –∏ –ø–æ–¥—Å—á–∏—Ç—ã–≤–∞–µ—Ç –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –±—É–∫–≤ –∏ —Ü–∏—Ñ—Ä.

# –ü—Ä–∏–º–µ—Ä –≤–≤–æ–¥–∞: hello world! 123
# –†–µ–∑—É–ª—å—Ç–∞—Ç:
# LETTERS 10
# DIGITS 3

st = "hello world! 123"
result = []

d = {"DIGITS": 0, "LETTERS": 0}
for c in st:
    if c.isdigit():
        d["DIGITS"] = +1
    elif c.isalpha():
        d["LETTERS"] += 1
print ("LETTERS", d["LETTERS"])
print ("DIGITS", d["DIGITS"])
```

```python
# –ó–ê–î–ê–ß–ê 14
# –£—Ä–æ–≤–µ–Ω—å: –°—Ä–µ–¥–Ω–∏–π

# –£—Å–ª–æ–≤–∏–µ:
# –ü—Ä–æ–≥—Ä–∞–º–º–∞ –ø—Ä–∏–Ω–∏–º–∞–µ—Ç —Å—Ç—Ä–æ–∫—É –∏ –ø–æ–¥—Å—á–∏—Ç—ã–≤–∞–µ—Ç –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –±—É–∫–≤ –≤ –≤–µ—Ä—Ö–Ω–µ–º –∏ –Ω–∏–∂–Ω–µ–º —Ä–µ–≥–∏—Å—Ç—Ä–µ.

# –ü—Ä–∏–º–µ—Ä –≤–≤–æ–¥–∞: Hello world!
# –†–µ–∑—É–ª—å—Ç–∞—Ç:
# UPPER CASE 1
# LOWER CASE 9

st = "Hello world!"
result = {"Upp": 0, "Low": 0}

for char in st:
  if char.isupper():
    result["Upp"] += 1
  elif char.islower():
    result["Low"] += 1
print(result)
```

```python
# –ó–ê–î–ê–ß–ê 15
# –£—Ä–æ–≤–µ–Ω—å: –°—Ä–µ–¥–Ω–∏–π

# –£—Å–ª–æ–≤–∏–µ:
# –í—ã—á–∏—Å–ª–∏—Ç—å –∑–Ω–∞—á–µ–Ω–∏–µ –≤—ã—Ä–∞–∂–µ–Ω–∏—è: a + aa + aaa + aaaa, –≥–¥–µ a ‚Äî –∑–∞–¥–∞–Ω–Ω–∞—è —Ü–∏—Ñ—Ä–∞.

# –ü—Ä–∏–º–µ—Ä –≤–≤–æ–¥–∞: 9
# –†–µ–∑—É–ª—å—Ç–∞—Ç: 11106

a = input()
n1 = int("%s" % a)
n2 = int("%s%s" % (a, a))
n3 = int("%s%s%s" % (a, a, a))
n4 = int("%s%s%s%s" % (a, a, a, a))
print (n1 + n2 + n3 + n4)
```
